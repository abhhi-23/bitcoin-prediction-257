```python
closedf = maindf[['Date','High','Low','Close']]
print("Shape of dataframe:", closedf.shape)
```

```python
closedf = closedf[closedf['Date'] > '2018-01-01']
close_stock = closedf.copy()
print("Total data for prediction: ",closedf.shape[0])
```

```python
fig = px.line(closedf, x=closedf.Date, y=closedf.Close,labels={'date':'Date','close':'Close Stock'})
fig.update_traces(marker_line_width=2, opacity=0.8, marker_line_color='orange')
fig.update_layout(title_text='Considered period to predict Bitcoin close price', 
                  plot_bgcolor='white', font_size=15, font_color='black')
fig.update_xaxes(showgrid=False)
fig.update_yaxes(showgrid=False)
fig.show()
```

```python
# deleting date column and normalizing using MinMax Scaler

del closedf['Date']
scaler = MinMaxScaler(feature_range=(0, 1))
closedf = scaler.fit_transform(closedf)
```

```python
closedf
```

```python
# keeping the training set as 60% and 40% testing set

training_size = int(len(closedf) * 0.60)
test_size = len(closedf) - training_size

train_data = closedf[:training_size, :]  
test_data = closedf[training_size:, :]   

print("train_data shape:", train_data.shape)
print("test_data shape:", test_data.shape)
```

```python
def create_dataset(dataset, time_step=1):
    dataX, dataY = [], []
    for i in range(len(dataset)-time_step-1):
        a = dataset[i:(i+time_step), :]  
        dataX.append(a)
        dataY.append(dataset[i + time_step, :])
    return np.array(dataX), np.array(dataY)
```

```python
time_step = 15
X_train, y_train = create_dataset(train_data, time_step)
X_test, y_test = create_dataset(test_data, time_step)

print("X_train: ", X_train.shape)
print("y_train: ", y_train.shape)
print("X_test: ", X_test.shape)
print("y_test", y_test.shape)
```

```python
X_train = X_train.reshape(X_train.shape[0], X_train.shape[1], X_train.shape[2])
X_test = X_test.reshape(X_test.shape[0], X_test.shape[1], X_test.shape[2])

print("X_train: ", X_train.shape)
print("X_test: ", X_test.shape)
```

```python
model = Sequential()
model.add(LSTM(units=10, input_shape=(time_step, train_data.shape[1]), activation="relu"))
model.add(Dense(train_data.shape[1])) 
model.compile(loss="mean_squared_error", optimizer="adam")
```

```python
history = model.fit(X_train,y_train,validation_data=(X_test,y_test),epochs=200,batch_size=32,verbose=1)
```

```python
train_predict=model.predict(X_train)
test_predict=model.predict(X_test)
train_predict.shape, test_predict.shape
```

```python

train_predict = scaler.inverse_transform(train_predict)
test_predict = scaler.inverse_transform(test_predict)
original_ytrain = scaler.inverse_transform(y_train)
original_ytest = scaler.inverse_transform(y_test)
```

```python
look_back=time_step
trainPredictPlot = np.empty_like(closedf)
trainPredictPlot[:, :] = np.nan
trainPredictPlot[look_back:len(train_predict)+look_back, :] = train_predict
print("Train predicted data: ", trainPredictPlot.shape)

# shift test predictions for plotting
testPredictPlot = np.empty_like(closedf)
testPredictPlot[:, :] = np.nan
testPredictPlot[len(train_predict)+(look_back*2)+1:len(closedf)-1, :] = test_predict
print("Test predicted data: ", testPredictPlot.shape)

names = cycle(['Original close price','Train predicted close price','Test predicted close price'])


plotdf = pd.DataFrame({'date': close_stock['Date'],
                       'original_close': close_stock['Close'].tolist(),
                       'train_predicted_close': trainPredictPlot[:, 0].tolist(),
                       'test_predicted_close': testPredictPlot[:, 0].tolist()})


fig = px.line(plotdf,x=plotdf['date'], y=[plotdf['original_close'],plotdf['train_predicted_close'],
                                          plotdf['test_predicted_close']],
              labels={'value':'Stock price','date': 'Date'})
fig.update_layout(title_text='Comparision between original close price vs predicted close price',
                  plot_bgcolor='white', font_size=15, font_color='black', legend_title_text='Close Price')
fig.for_each_trace(lambda t:  t.update(name = next(names)))

fig.update_xaxes(showgrid=False)
fig.update_yaxes(showgrid=False)
fig.show()
```

```python
import numpy as np

# Get the last sequence from the test data for prediction
x_input = test_data[len(test_data) - time_step:]
temp_input = list(x_input)
temp_input = [item for sublist in temp_input for item in sublist] 


lst_output_high = []
lst_output_low = []
lst_output_close = []
n_steps = time_step
pred_days = 7
i = 0

while i < pred_days:
    if len(temp_input) >= time_step * test_data.shape[1]:
        x_input = np.array(temp_input[-time_step * test_data.shape[1]:]).reshape((1, n_steps, test_data.shape[1]))
        
        # Predict next values
        yhat = model.predict(x_input, verbose=0)
        
        # Update temp_input with new predictions
        temp_input.extend(yhat[0].tolist())
        
        # Append predictions to the respective lists
        lst_output_high.append(yhat[0][0])  # High price
        lst_output_low.append(yhat[0][1])   # Low price
        lst_output_close.append(yhat[0][2]) # Close price
        
        i += 1
    else:
        break

# Prepare the data for inverse transformation
predicted_data = np.array([lst_output_high, lst_output_low, lst_output_close]).T

# Inverse transform the predictions to original scale
predicted_data = scaler.inverse_transform(predicted_data)

# Separate the high, low, and close prices
lst_output_high = predicted_data[:, 0]
lst_output_low = predicted_data[:, 1]
lst_output_close = predicted_data[:, 2]

print("Predicted High Prices for the next 7 days: ", lst_output_high)
print("Predicted Low Prices for the next 7 days: ", lst_output_low)
print("Predicted Close Prices for the next 7 days: ", lst_output_close)

# Calculate the highest, lowest, adn average closing prices for the next 7 days
highest_price_next_7_days = max(lst_output_high)
lowest_price_next_7_days = min(lst_output_low)
average_close_price_next_7_days = np.mean(lst_output_close)

print("Highest price for the next 7 days: ", highest_price_next_7_days)
print("Lowest price for the next 7 days: ", lowest_price_next_7_days)
print("Average closing price for the next 7 days: ", average_close_price_next_7_days)
```
